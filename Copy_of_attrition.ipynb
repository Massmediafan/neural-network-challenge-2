## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers
import tensorflow as tf
from tensorflow.keras import backend as K
from tensorflow.keras.layers import Layer

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head(100)
```

```python
attrition_counts = attrition_df["Attrition"].value_counts()
print(attrition_counts)
```

```python
department_counts = attrition_df["Department"].value_counts()
print(department_counts)
```

```python
K.clear_session()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns

y_df = attrition_df[["Attrition", "Department"]]
```

```python
# Create a list of at least 10 column names to use as X data
x_columns = [
    "Age",
    "BusinessTravel",
    "DistanceFromHome",
    "Education",
    "EnvironmentSatisfaction",
    "JobInvolvement",
    "StockOptionLevel",
    "TotalWorkingYears",
    "WorkLifeBalance",
    "YearsAtCompany"
]


# Create X_df using your selected columns
X_df = attrition_df[x_columns]

# Show the data types for X_df
X_df.dtypes
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
non_numeric_columns = X_df.select_dtypes(include=["object"]).columns
print("Non-numeric columns:", non_numeric_columns)

X_df_encoded = pd.get_dummies(X_df, columns=non_numeric_columns, drop_first=True)
X_df = X_df_encoded
X_df.dtypes
```

```python
X_df = X_df.astype({col: 'int' for col in X_df.select_dtypes(include=['bool']).columns})
X_df.dtypes
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_df, y_df, test_size=0.2, random_state=42)
```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
scaler.fit(X_train)

# Scale the training and testing data
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns, index=X_train.index)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns, index=X_test.index)


X_train_scaled.head()
```

```python
# Create a OneHotEncoder for the Department column
from sklearn.preprocessing import OneHotEncoder

department_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore', drop='first') # Added handle_unknown='ignore' to handle unseen values during prediction

# Fit the encoder to the training data
department_encoder.fit(y_train[["Department"]])

# Transform the training and testing data using the fitted encoder
# Removed .toarray() as it's unnecessary when sparse=False
department_train_encoded = department_encoder.transform(y_train[["Department"]])
department_test_encoded = department_encoder.transform(y_test[["Department"]])

# Create DataFrames for clear viewing
department_train_encoded_df = pd.DataFrame(
    department_train_encoded,
    columns=department_encoder.get_feature_names_out(["Department"]),
    index=y_train.index
)
department_test_encoded_df = pd.DataFrame(
    department_test_encoded,
    columns=department_encoder.get_feature_names_out(["Department"]),
    index=y_test.index
)

department_train_encoded.shape
# department_test_encoded.shape
```

```python
# Create a OneHotEncoder for the Attrition column
attrition_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore', drop='first')

# Fit the encoder to the training data
attrition_encoder.fit(y_train[["Attrition"]])

# Transform the training and testing data using the fitted encoder
attrition_train_encoded = attrition_encoder.transform(y_train[["Attrition"]])
attrition_test_encoded = attrition_encoder.transform(y_test[["Attrition"]])

# Create DataFrames for clear viewing
attrition_train_encoded_df = pd.DataFrame(
    attrition_train_encoded,
    columns=attrition_encoder.get_feature_names_out(["Attrition"]),
    index=y_train.index
)
attrition_test_encoded_df = pd.DataFrame(
    attrition_test_encoded,
    columns=attrition_encoder.get_feature_names_out(["Attrition"]),
    index=y_test.index
)

print(attrition_train_encoded.shape)
print(attrition_test_encoded.shape)
```

```python
# Find the number of columns in the X training data
input_dim = X_train_scaled.shape[1]
input_dim
```

```python
class NamedDense(Layer):
    def __init__(self, units, activation, name):
        super(NamedDense, self).__init__(name=name)
        self.dense = layers.Dense(units, activation=activation)

    def call(self, inputs):
        return self.dense(inputs)
```

```python
# Create the input layer
input_layer = layers.Input(shape=(11,))


# Create at least two shared layers
shared_layer_1 = layers.Dense(64, activation='relu')(input_layer)
shared_layer_2 = layers.Dense(32, activation='relu')(shared_layer_1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
department_hidden = layers.Dense(16, activation='relu')(shared_layer_2)

# Create the output layer
department_output = NamedDense(2, activation='softmax', name="department_output")(department_hidden)

```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attrition_hidden = layers.Dense(16, activation='relu')(shared_layer_2)

# Create the output layer
attrition_output = NamedDense(1, activation='sigmoid', name="attrition_output")(attrition_hidden)
```

```python
print(department_output)  # Should print a tensor with name 'department_output'
print(attrition_output)   # Should print a tensor with name 'attrition_output'
```

```python
model = Model(
    inputs=input_layer,
    outputs={
        "department_output": department_output,
        "attrition_output": attrition_output
    }
)


print(list(model.output.keys()))
```

```python
# Compile the model
model.compile(
    optimizer="adam",
    loss={
        "department_output": "categorical_crossentropy",  # Multi-class loss
        "attrition_output": "binary_crossentropy",       # Binary loss
    },
    metrics={
        "department_output": ["accuracy"],
        "attrition_output": ["accuracy"],
    }
)

# Summarize the model
model.summary()
```

```python
try:
    history = model.fit(
        X_train_scaled,
        {
            "department_output": department_train_encoded,  # Matches layer name
            "attrition_output": attrition_train_encoded,    # Matches layer name
        },
        epochs=100,  # Run 1 epoch to debug
        batch_size=32,
        validation_data=(
            X_test_scaled,
            {
                "department_output": department_test_encoded,  # Matches layer name
                "attrition_output": attrition_test_encoded,    # Matches layer name
            }
        ),
        verbose=1
    )
    print("Training completed successfully.")
except Exception as e:
    print("Error during training:", e)
```

```python
# Evaluate the model with the testing data
results = model.evaluate(
    X_test_scaled,
    {
        "department_output": department_test_encoded,
        "attrition_output": attrition_test_encoded,
    },
    verbose=1
)

# Print results
results
```

```python
# Print the accuracy for both department and attrition
department_accuracy = results[3]
attrition_accuracy = results[4]

print(f"Department Accuracy: {department_accuracy}")
print(f"Attrition Accuracy: {attrition_accuracy}")
```
